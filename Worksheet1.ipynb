```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
myRGBMatrix = np.zeros([100,200,3], dtype='uint8')
blue = [0, 0, 255]
orange = [255, 127, 0]

print(myRGBMatrix.shape)
print('Matrix type:\n {}'.format(myRGBMatrix.dtype))

myRGBMatrix[:, :(myRGBMatrix.shape[1]//2), :] = orange
myRGBMatrix[:, (myRGBMatrix.shape[1]//2):, :] = blue

print(myRGBMatrix.shape[1]//2)

plt.imshow(myRGBMatrix)
plt.show()
```

```python
#  Create a 10x10 Greyscale matrix and then change all pixels below 100 to black

myMatrix = np.zeros([10,10], dtype='uint8')

print(myMatrix.shape)
print('Matrix type:\n {}'.format(myMatrix.dtype))

myMatrix = np.random.randint(0, 255, size=myMatrix.shape)

plt.imshow(myMatrix, cmap='gray', vmin=0, vmax=255)
plt.show()

myMatrix[myMatrix<100]=0
plt.imshow(myMatrix, cmap='gray', vmin=0, vmax=255)
plt.show()  


MyBinaryMatrix = np.where(myMatrix<127, 0, 255)
plt.imshow(MyBinaryMatrix, cmap='gray', vmin=0, vmax=255)
plt.show()  

# research np.where
```

```python
# Save image to disk


myRGBMatrix = np.zeros([100,200,3], dtype='uint8')
blue = [0, 0, 255]
orange = [255, 127, 0]

print(myRGBMatrix.shape)
print('Matrix type:\n {}'.format(myRGBMatrix.dtype))

myRGBMatrix[:, :(myRGBMatrix.shape[1]//2), :] = orange
myRGBMatrix[:, (myRGBMatrix.shape[1]//2):, :] = blue

print(myRGBMatrix.shape[1]//2)

plt.imshow(myRGBMatrix)
plt.show()

# save matrix as an image to disk using PIL
Image.fromarray(myRGBMatrix.astype(np.uint8)).save('random.png')

```
