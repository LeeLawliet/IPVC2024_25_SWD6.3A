```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

```python
# Task 1
# Separating Red, Blue, and Green channels
imBGR = cv2.imread('./img/goldfish.jpg') # BGR
imRGB = imBGR[:, :, ::-1] # Conversion to RGB
imRed = imRGB.copy()
imRed[:, :, [1,2]] = 0
imBlue = imRGB.copy()
imBlue[:, :, [0,2]] = 0
imGreen = imRGB.copy()
imGreen[:, :, [0,1]] = 0

# imBlue = imRGB[:, :, 1]
# imGreen = imRGB[:, :, 2]
print(imBGR.shape)
print(imRGB.shape)
plt.imshow(imRGB, vmin=0, vmax=255)
plt.show()
plt.imshow(imRed, vmin=0, vmax=255)
plt.show()
plt.imshow(imBlue, vmin=0, vmax=255)
plt.show()
plt.imshow(imGreen, vmin=0, vmax=255)
plt.show()
```

```python
#Task 2
#Increase brightness
imBGR = cv2.imread('./img/lena.jpg') # BGR
imRGB = imBGR[:, :, ::-1] # Conversion to RGB
imBright = imRGB.copy()

for rowPix in range(imBright.shape[0]): # Loop over rows
    for colPix in range(imBright.shape[1]):  # Loop over columns
        for channel in range(3):  # Loop over RGB channels
            newPix = (imBright[rowPix, colPix, channel]).astype("uint16") + 50 # conver to uint16 so value can be >255
            if newPix > 255:
                print(newPix)
                imBright[rowPix, colPix, channel] = 255
            else:
                imBright[rowPix, colPix, channel] = newPix

plt.imshow(imRGB)
plt.show()
plt.imshow(imBright)
plt.show()
```

```python
#Task 3
#Image Trimming
imBGR = cv2.imread('./img/lena.jpg') # BGR
imRGB = imBGR[:, :, ::-1] # Conversion to RGB
imTrimmed = imRGB[10:-10, 10:-10, :]

plt.imshow(imRGB, vmin=0, vmax=255)
plt.show()

plt.imshow(imTrimmed, vmin=0, vmax=255)
plt.show()
```

```python
#Task 4
#Image Flipping
imBGR = cv2.imread('./img/lena.jpg') # BGR
imRGB = imBGR[:, :, ::-1] # Conversion to RGB
imHorizontal = imRGB[::-1, :, :]
imVertical = imRGB[:, ::-1, :]

plt.imshow(imRGB, vmin=0, vmax=255)
plt.show()

plt.imshow(imHorizontal, vmin=0, vmax=255)
plt.show()

plt.imshow(imVertical, vmin=0, vmax=255)
plt.show()
```
