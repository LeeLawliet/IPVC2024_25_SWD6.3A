```python
import numpy as np
import matplotlib.pyplot as plt
import math

x = np.array(['Jan', 'Feb', 'Mar', 'Apr'])
y = np.array([45, 78, 34, 23])

plt.bar(x,y)
plt.xlabel('Months')
plt.ylabel('Rainfall (mm)')
plt.title('Rainfall')
plt.show()
```

```python
from matplotlib import pyplot as plt 
import numpy as np 
import cv2

imgGray = cv2.imread("./img/Fog.jpg",0)

bins=256

flattened_imgGray = imgGray.ravel()

plt.hist(flattened_imgGray,bins, range=[0,255])
plt.xlabel("Intensity") 
plt.ylabel("No of pixels") 
plt.title('Image Histogram') 
plt.show() 

plt.imshow(imgGray, cmap='gray', vmin=0, vmax=255)
plt.show()
```
