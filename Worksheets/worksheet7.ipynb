```python
# Task 1

from matplotlib import pyplot as plt 
import numpy as np 
import cv2

imBGR = cv2.imread("../img/lena.jpg")
imRGB = imBGR[:, :, ::-1] 
imGray = cv2.cvtColor(imRGB, cv2.COLOR_RGB2GRAY)

bins=256

flattened_imgGray = imGray.ravel()

fig, ax = plt.subplots(2, 2, figsize=(12, 12))

ax[0, 0].hist(flattened_imgGray, bins, range=[0, 255])
ax[0, 0].set_xlabel("Intensity")
ax[0, 0].set_ylabel("Number of Pixels")
ax[0, 0].set_title('Image Histogram')

ax[0, 1].imshow(imGray, cmap='gray')
ax[0, 1].set_title('Lena')

# RGB Histogram

bins=256

imgRed = imRGB[:, :, 0]
imgGreen = imRGB[:, :, 1]
imgBlue = imRGB[:, :, 2]

ax[1, 0].hist(imgRed.ravel(), bins, color='red', histtype='step', range=[0, 255])
ax[1, 0].hist(imgGreen.ravel(), bins, color='green',  histtype='step', range=[0, 255])
ax[1, 0].hist(imgBlue.ravel(), bins, color='blue',  histtype='step', range=[0, 255])
ax[1, 0].set_xlabel("Intensity")
ax[1, 0].set_ylabel("Number of Pixels")
ax[1, 0].set_title('Image Histogram')

ax[1, 1].imshow(imRGB)
ax[1, 1].set_title('Lena RGB')
plt.show()
```
