```python
# Task 1

from matplotlib import pyplot as plt 
import numpy as np 
import cv2

imBGR = cv2.imread("../img/lena.jpg")
imRGB = imBGR[:, :, ::-1] 
imGray = cv2.cvtColor(imRGB, cv2.COLOR_RGB2GRAY)

bins=256

flattened_imgGray = imGray.ravel()

fig, ax = plt.subplots(2, 2, figsize=(12, 12))

ax[0, 0].hist(flattened_imgGray, bins, range=[0, 255])
ax[0, 0].set_xlabel("Intensity")
ax[0, 0].set_ylabel("Number of Pixels")
ax[0, 0].set_title('Image Histogram')

ax[0, 1].imshow(imGray, cmap='gray')
ax[0, 1].set_title('Lena')

# RGB Histogram

bins=256

imgRed = imRGB[:, :, 0]
imgGreen = imRGB[:, :, 1]
imgBlue = imRGB[:, :, 2]

ax[1, 0].hist(imgRed.ravel(), bins, color='red', histtype='step', range=[0, 255])
ax[1, 0].hist(imgGreen.ravel(), bins, color='green',  histtype='step', range=[0, 255])
ax[1, 0].hist(imgBlue.ravel(), bins, color='blue',  histtype='step', range=[0, 255])
ax[1, 0].set_xlabel("Intensity")
ax[1, 0].set_ylabel("Number of Pixels")
ax[1, 0].set_title('Image Histogram')

ax[1, 1].imshow(imRGB)
ax[1, 1].set_title('Lena RGB')
plt.show()
```

```python
# Shift pixels by one

import cv2
import numpy as np
from matplotlib import pyplot as plt

img = np.random.randint(0,255, size=[20,20, 3], dtype="uint8")
kernel = np.array([[0,0,0], [1,0,0], [0,0,0]])

plt.imshow(img)
plt.title('Original')
plt.show()

dst = cv2.filter2D(img,-1,kernel)

plt.imshow(dst)
plt.show()
```

```python
# Gaussian filter

import cv2
import numpy as np
from matplotlib import pyplot as plt

imgBGR = cv2.imread('../img/lena.jpg')
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)

kernel = np.array([[1,2,1], [2,4,2], [1,2,1]]) / 16

plt.figure(figsize=(20,20))
plt.imshow(imgRGB)
plt.title('Original')
plt.show()

dst = cv2.filter2D(imgRGB,-1,kernel)

plt.figure(figsize=(20,20))
plt.imshow(dst)
plt.show()

# with cv2.gaussianblur 

dst2 = cv2.GaussianBlur(imgRGB, ksize = (3,3), sigmaX=1, sigmaY=1)
plt.imshow(dst2)
plt.show()
```
