```python
# Task 1

from matplotlib import pyplot as plt 
import numpy as np 
import cv2

imgGray = cv2.imread("../img/lena.jpg",0)

bins=256

flattened_imgGray = imgGray.ravel()

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].hist(flattened_imgGray, bins, range=[0, 255])
ax[0].set_xlabel("Intensity")
ax[0].set_ylabel("Number of Pixels")
ax[0].set_title('Image Histogram')

ax[1].imshow(imgGray, cmap='gray', vmin=0, vmax=255)
ax[1].set_title('Grayscale Image')
plt.show()
```

```python
# Task 2

from matplotlib import pyplot as plt 
import numpy as np 
import cv2

imgDots = cv2.imread("../img/blackdots.png")

rows, cols, channels = imgDots.shape
blackCount = 0

for row in range(rows):
    for col in range(cols):
        if np.array_equal(imgDots[row, col], [0, 0, 0]): 
            cv2.circle(imgDots, center=(col,row), radius=10, color=(255,0,0), thickness=3)
            blackCount+=1

print(blackCount)

label = 'No of pixels: ' + str(blackCount)
cv2.putText(imgDots, label, (10,30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0), 2)


plt.imshow(imgDots)
plt.show()
```

```python
# Task 3
import random

randImg = []

i = 0
for i in range (15 * 15):
    randRGB = []
    j = 0

    while (j < 3):
        color = (random.randint(0, 1)) * 255
        randRGB.append(color)
        j+=1
    
    randImg.append(randRGB)
    i+=1

print (randImg)

randImg = np.array(randImg, dtype=np.uint8).reshape(15, 15, 3)

plt.imshow(randImg, vmin=0, vmax=255)
plt.show()

# circling

upscaledImage = cv2.resize(randImg, (150, 150), interpolation=cv2.INTER_NEAREST)

rows, cols, channels = randImg.shape
rCount = 0
gCount = 0
bCount = 0

for row in range(rows):
    for col in range(cols):
        if np.array_equal(randImg[row, col], [255, 0, 0]): 
            cv2.circle(upscaledImage, center=(col*10+5,row*10+5), radius=7, color=(255,0,0), thickness=1)
            rCount+=1
        elif(np.array_equal(randImg[row, col], [0, 255, 0])):
            cv2.circle(upscaledImage, center=(col*10+5,row*10+5), radius=7, color=(0,255,0), thickness=1)
            gCount+=1
        elif(np.array_equal(randImg[row, col], [0, 0, 255])):
            cv2.circle(upscaledImage, center=(col*10+5,row*10+5), radius=7, color=(0,0,255), thickness=1)
            bCount+=1

print(f"red: {rCount}\ngreen: {gCount}\nblue: {bCount}")

plt.imshow(upscaledImage)
plt.show()
```
