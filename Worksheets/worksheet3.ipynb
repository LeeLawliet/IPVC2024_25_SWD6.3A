```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

```python
imBGR = cv2.imread('./img/goldfish.jpg') # BGR
imRGB = imBGR[:, :, ::-1] # Conversion to RGB

plt.imshow(imRGB, vmin=0, vmax=255)
plt.show()

plt.figure(figsize=(20,20)) # figsize uses inches for measurement
plt.subplot(1, 2, 1)
plt.imshow(imRGB, vmin=0, vmax=255)

plt.subplot(1, 2, 2)
plt.imshow(imRGB, vmin=0, vmax=255)

plt.show()
```

```python
#Task 2
imBGR = cv2.imread('./img/goldfish.jpg') # BGR
imRGB = imBGR[:, :, ::-1] # Conversion to RGB
imRed = imRGB.copy()
imRed[:, :, [1,2]] = 0
imBlue = imRGB.copy()
imBlue[:, :, [0,1]] = 0
imGreen = imRGB.copy()
imGreen[:, :, [0,2]] = 0

plt.figure(figsize=(10,10)) # figsize uses inches for measurement

plt.subplot(2, 2, 1)
plt.imshow(imRGB, vmin=0, vmax=255)
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(imRed, vmin=0, vmax=255)
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(imBlue, vmin=0, vmax=255)
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(imGreen, vmin=0, vmax=255)
plt.axis('off')

plt.show()

Image.fromarray(imRed.astype(np.uint8)).save('red.png')
Image.fromarray(imBlue.astype(np.uint8)).save('blue.png')
Image.fromarray(imGreen.astype(np.uint8)).save('green.png')
```

```python
#Task 3
imBGR = cv2.imread('./img/goldfish.jpg') # BGR
imRGB = imBGR[:, :, ::-1] # Conversion to RGB
imHeight = imRGB.shape[0]
imWidth = imRGB.shape[1]
print(imHeight)
print(imWidth)

centerHeight = imHeight // 2
centerWidth = imWidth // 2

topLeft = imRGB[:centerHeight, :centerWidth, :]
topRight = imRGB[:centerHeight, centerWidth:, :]
bottomLeft = imRGB[centerHeight:, centerWidth:, :]
bottomRight = imRGB[centerHeight:, :centerWidth:, :]

plt.imshow(imRGB)
plt.show

plt.figure(figsize=(10,5)) # figsize uses inches for measurement

plt.subplot(2,2,1)
plt.imshow(topLeft)

plt.subplot(2,2,2)
plt.imshow(topRight)

plt.subplot(2,2,3)
plt.imshow(bottomLeft)

plt.subplot(2,2,4)
plt.imshow(bottomRight)
plt.show()

```
