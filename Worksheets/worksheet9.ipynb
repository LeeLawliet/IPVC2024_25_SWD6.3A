```python
# Task 1

import imutils as im
import cv2
import numpy as np
from matplotlib import pyplot as plt

imgBGR = cv2.imread('../img/lena.jpg')
imgGray = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2GRAY)

resized1 = cv2.resize(imgGray, (400,400))

rotated = im.rotate(resized1, 30)
rotate_bound = im.rotate_bound(resized1, 30)

plt.imshow(rotated, cmap='gray')
plt.title("Rotated")
plt.show()

plt.imshow(rotate_bound, cmap='gray')
plt.title("Rotate Bound")
plt.show()

# The difference between Rotate and Rotate Bound:
# Rotate will take the exact size of the image and rotate it. since the image size stays the same,
# the image will go outside the border of the image.
# Rotate bound will resize the image inside to stay within the borders of the image
```

```python
# Task 2

import imutils as im
import cv2
import numpy as np
from matplotlib import pyplot as plt

imgBGR = cv2.imread('../img/lena.jpg')
imgGray = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2GRAY)

resized1 = cv2.resize(imgGray, (400,400))
image1Dimensions = np.shape(resized1)

rotated = im.rotate(resized1, 30)
rotated =  cv2.resize(rotated, (image1Dimensions[1], image1Dimensions[0])) # Image1Dimesions[1] -> X | Image1Dimensions[0] -> Y
rotate_bound = im.rotate_bound(resized1, 30)
rotate_bound =  cv2.resize(rotate_bound, (image1Dimensions[1], image1Dimensions[0])) # Image1Dimesions[1] -> X | Image1Dimensions[0] -> Y

userInput = input("Enter a whole number to represent how transparent you wish the first image to be:")
blended1 = cv2.addWeighted(resized1, (int(userInput)/100), rotated, ((100-int(userInput))/100), 0) # Only works when both images are of the same size
blended2 = cv2.addWeighted(resized1, (int(userInput)/100), rotate_bound, ((100-int(userInput))/100), 0) # Only works when both images are of the same size

plt.imshow(blended1, cmap='gray')
plt.title("Rotated")
plt.show()

plt.imshow(blended2, cmap='gray')
plt.title("Rotate Bound")
plt.show()
```

```python
# Task 4
import imutils as im
import cv2
import numpy as np
from matplotlib import pyplot as plt

imgBGR1 = cv2.imread('../img/monkeydark.jpeg')
imgBGR2 = cv2.imread('../img/lena_square.png')

imgGray1 = cv2.cvtColor(imgBGR1, cv2.COLOR_BGR2GRAY)
imgGray2 = cv2.cvtColor(imgBGR2, cv2.COLOR_BGR2GRAY)

image1Dimensions = np.shape(imgGray1)
image2Dimensions = np.shape(imgGray2)

if (image1Dimensions[0] > image2Dimensions[0]):
    smallerImage = imgGray2
    resizedImage = cv2.resize(imgGray1, (image2Dimensions[1],image2Dimensions[0]))

if (image2Dimensions[0] > image1Dimensions[0]):
    smallerImage = imgGray1
    resizedImage = cv2.resize(imgGray2, (image1Dimensions[1],image1Dimensions[0]))

stitchedImages = np.concatenate((smallerImage,resizedImage), axis=1) # Stitch images
plt.imshow(stitchedImages, cmap='gray')
plt.title("Stitched 1")
plt.show()
```
